## Creating Model
We'll be using KNearestNeighbors to calculate the distance between products using cosine similarity metric. 

#### Item Similarity

```python
from sklearn.neighbors import NearestNeighbors
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(mat.values)
distances, indices = knn.kneighbors(mat.values, n_neighbors=3)
```

Parameter for the number of the nearest neighbors is set to be 3.

```python
indices
```

**indices** shows the nearest product to each product.. Each row corresponds to row in `mat`. The first element in a row is the most similar(nearest) product. It is the product itself. The second element is the second nearest, and the third element is the third nearest. For example, in the first row **[0, 1662, 2]**, the nearest product to 'product[0]' is itself, the second nearest is 'product[1662]' and the third nearest is 'product[2]'.

```python
distances
```

**distances** shows the distance between these products. Each number in this array corresponds to the number in the **indices** array. 

### Reccomend Similar Product to a product

```python
def recommend_product(product):
    index_user_likes = mat.index.tolist().index(product) # get an index for that product
    sim_products = indices[index_user_likes].tolist() # make list of similar products
    product_distances = distances[index_user_likes].tolist() # list of distances to similar products
    id_product = sim_products.index(index_user_likes) # get positition of the product itself in indices and distances
    
    print('Similar Products to '+str(mat.index[index_user_likes])+': \n')
    
    sim_products.remove(index_user_likes) # remove the product itself in indices
    product_distances.pop(id_product) #remove the movie itself in distances
    
    j = 1
    
    for i in sim_products:
        print(str(j)+': '+str(mat.index[i])+', the distance with '+str(product)+': '+str(product_distances[j-1]))
        j+=1
```

## Recommend A Product for a Selected User
Instead of recommending products based on a selected product, I will now be creating a recommender which recommends products to the user based on predicting ratings for products the user has not reviewed yet. 

Algorithm for this recommendation engine is as follows:
1. Calculate similarity between products using KNN (Similarities are calculated based on all ratings by all users. We did this previously)
1. For each product a user has not reviewed, predict the rating they would give:
    * Find closest product based on similarity calculated in step 1
    * Find closest product, calculate the weighted average of ratings by the user. The inverse distance metric is used as the weight.
    * Use the weighted average of the ratings as the predicted rating for the product by the user.
1. Recommend products which have the highest predicted ratings for the user.

### Predict Rating for a product by a user
In this part we will practice how to predict a rating for a product by a user. For example, `customer_unique_id = 0000366f3b9a7992bf8c76cfdf3221e2` has not given a rating for `product_id = 001b72dfd63e9833e8c02742adf472e3`. Let's try and predict what this rating would be.

First thing we do is to find the nearest neighbors for this product using **KNN(n_neighbors = 3).**

```python
mat
```

```python
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(mat.values)
distances, indices = knn.kneighbors(mat.values, n_neighbors=3)
```

```python
index_for_product = mat.index.tolist().index('001b72dfd63e9833e8c02742adf472e3') #return a 0 (first product in matrix)
sim_products = indices[index_for_product].tolist() # list of similar productss
product_distances = distances[index_for_product].tolist() # list of distances to similar products
id_product = sim_products.index(index_for_product) #get position of product itself
print(sim_products, product_distances)
sim_products.remove(index_for_product) # remove products itself in indices
product_distances.pop(id_product) # remove product itself in distances

print('The Nearest Product to 001b72dfd63e9833e8c02742adf472e3: ', sim_products)
print('The Distance from 001b72dfd63e9833e8c02742adf472e3:', product_distances)
```

#### Predict a Rating

##### Formula
Formula to calculate the predicted rating is as follows:
$$ \hat{R}_{pu} = \frac{\sum_j R_{ju}S_{pj}}{\sum_j S_{pj}}$$
Where $\hat{R}_{pu}$ is the predicted rating for a product , $p$, by user, $u$. $R_{ju}$ is the actual rating for product, $j$, by user, $u$. $S_{pj}$ is the similarity between product, $p$, and product , $j$.

The formula simply implies that the predicted rating for a product is the weighted average of ratings for similar products. The weight for each rating, $\frac{S_{pj}}{\sum_kS_{pk}}$ becomes greater when product $p$ and product $j$ are more similar. The denominator of this term makes the sum of all the weights in $\hat{R}_{pu}$ become 1.

##### Application

```python
product_similarity = [-x+1 for x in product_distances] # inverse distance

predicted_rating = (product_similarity[0]*mat.iloc[sim_products[0],0] + \
                    product_similarity[1]*mat.iloc[sim_products[1],0]) / sum(product_similarity)
print(predicted_rating)
```

### Building Recommnder

```python
mat
```

```python
mat1 = mat.copy() # make a copy to adjust in function
```

```python
def recommend_products(user, num_recommended_products):
    
    user = str(user) # make sure the user id is a string
    
    print(f"The list of the Products User: {user} has bought/reviewed \n")
    
    for prod in mat[mat[user] > 0.0][user].index.tolist(): # loop through all products that have review
        print(prod)
    
    print('\n')
    
    recommended_products = [] # initialize list 
    
    for prod in mat[mat[user]==0.0].index.tolist(): #loop through all unreviewed products
        index_df = mat.index.tolist().index(prod)
        predicted_rating = mat1.iloc[index_df, mat1.columns.tolist().index(user)]
        recommended_products.append((prod, predicted_rating))
    
    sorted_rm = sorted(recommended_products, key=lambda x:x[1], reverse=True)
    
    print('The list of the Recommended Products \n')
    rank = 1 
    for recommended_product in sorted_rm[:num_recommended_products]:
        print(f'{rank}: {recommended_product[0]} - predicted rating: {recommended_product[1]}')
        rank += 1
```

```python
def product_recommender(user, num_neighbors, num_recommendation):
    knn = NearestNeighbors(metric='cosine', algorithm='brute')
    knn.fit(mat.values)
    distances, indices = knn.kneighbors(mat.values, n_neighbors=num_neighbors)
    
    user_index = mat.columns.tolist().index(user)
    
    for i, prod in list(enumerate(mat.index)):
        # if user hasnt given this product a rating:
        if mat.iloc[i, user_index] == 0.0: 
            sim_prods = indices[i].tolist() # add that product's similar products to similar product list
            prod_distances = distances[i].tolist() # add distances to sim. prod. to list
            
            if i in sim_prods: #removing itself from lists
                id_prod = sim_prods.index(i)
                sim_prods.remove(i)
                prod_distances.pop(id_prod)
                
            else:
                sim_prods = sim_prods[:num_neighbors-1]
                prod_distances = prod_distances[:num_neighbors-1]
            
            prod_similarity = [1-x for x in prod_distances] # inverse distance
            prod_similarity_copy = prod_similarity.copy() # create a copy
            nominator = 0
            
            for s in range(0, len(prod_similarity)):
                if mat.iloc[sim_prods[s], user_index] == 0.0: #check if the similar product has rating of 0 for user
                    if len(prod_similarity_copy) == (num_neighbors - 1):
                        prod_similarity_copy.pop(s)
                        
                    else:
                        prod_similarity_copy.pop(s-(len(prod_similarity)-len(prod_similarity_copy)))
                    
                else: # similar prod has rating
                    nominator = nominator + prod_similarity[s]*mat.iloc[sim_prods[s], user_index]
                    
            if len(prod_similarity_copy) > 0: # make sure there are similar prods
                if sum(prod_similarity_copy) > 0: # Don't want to divide by 0!
                    predicted_r = nominator/sum(prod_similarity_copy) # calculated the predicted rating
                
                else:
                    predicted_r = 0.0
            else:
                predicted_r = 0.0
            
            mat1.iloc[i, user_index] = predicted_r # enter the predicted rating into the copy matrix
        
    recommend_products(user, num_recommendation)
```

```python
import random
user = random.choice(mat.columns.tolist())
product_recommender(user, 10, 10)
```
